# Powers over Generations 

In our analysis of Pokemon stats
We can see the trends of various powers vary over the 6 Generations.

# Line plot 

```python
#setting the plot size
fig = plt.figure(figsize=(12,9))
for i in data.columns[5:10]:   
    generations= data.groupby("Generation")[i].mean().index
    powers = data.groupby("Generation")[i].mean().values
    plt.plot(generations,powers)
plt.legend(data.columns[5:10], loc='upper left')
plt.ylabel("Average power points", color = 'black')
plt.xlabel("Pokemon Generations", color = 'black')
plt.title("Pokemon Powers by Generations",color='black')
plt.show()

```

```python
THe above plot gives us the overview of the avg power points for [attack, defense, specialattack,hp ] 
for all the pokemons over all the 6 Generations. We can see how the trends in powers vary.
The reason to choose line chart is that is shows the overall trend wheter it is increasing or decreasing.
As from the plot we can clearly interpret that Gen 2 has a low attack power compared to the other Generations.
And also Gen1 has low defense power compared to the other Generations.


The type of power[3 rd dimension] is encoded using the colour attribute, as the plot represents 
different colours for different powers
```

# Other encodings



The same data can also be visualized using a scatter plot with different colors indicating different powers.
But this encoding is not chosen as it could not be so clear to interpret the highest powers for each generation 
and is also difficult to read compared to the above plots.

The above data can also be encoded using the pie charts for each power. Each section would be a proportion of powers 
per each generation. For this we need to plot 5 different pie charts , 
seperatly for each power which is not so reader friendly.


While the bar chart encodes all the information into a single plot, I would prefer using a line chart as it describes
the overall trend of how the  powers vary over the generations which is easily readable.
